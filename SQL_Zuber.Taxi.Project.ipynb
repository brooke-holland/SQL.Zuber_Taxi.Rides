## Test the hypothesis:

*"The average duration of rides from the Loop to O'Hare International Airport changes on rainy Saturdays."*

- The null hypothesis (H0) assumes no differences 
- Alternative hypothesis (H1) proposes a difference

```python
# importing the new dataset given after collecting the data from Step 4. 

df7 = pd.read_csv('/datasets/project_sql_result_07.csv')
```

```python
#converting datasets, column names to lowercase

df7.columns = df7.columns.str.lower()
```

```python
#printing summary of df1 data type info

df7.info()
```

```python
# changing to datetime 
df7['start_ts'] = pd.to_datetime(df7['start_ts'], format='%Y-%m-%d %H:%M:%S')

df7.head()
```

```python
# testing two independent samples from descriptive stats
from scipy.stats import ttest_ind

# Separate the data
rainy_saturdays = df7[(df7['weather_conditions'] == 'Bad') & (df7['start_ts'].dt.dayofweek == 5)]['duration_seconds']
non_rainy = df7[(df7['weather_conditions'] == 'Good') & (df7['start_ts'].dt.dayofweek == 5)]['duration_seconds']

# Set significance level
alpha = 0.05

# Perform t-test
t_stat, p_value = ttest_ind(rainy_saturdays, non_rainy, nan_policy='omit')
result = ttest_ind(rainy_saturdays, non_rainy)

# Interpret the results
if p_value < alpha:
    print("Reject the null hypothesis. There is a significant difference in ride duration on rainy Saturdays.")
else:
    print("Fail to reject the null hypothesis. No significant difference in ride duration on rainy Saturdays.")
    
print(result.pvalue)
```
